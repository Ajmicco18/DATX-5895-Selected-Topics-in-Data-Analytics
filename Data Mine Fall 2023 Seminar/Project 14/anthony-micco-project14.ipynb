## Question 2

My favorite function that we worked with this semester was the tapply function. I thought it was really neat how it could group the data and perform a calucation simply based on what we inputted into the function. It also made it extremely simple to do calculations that may have required loops in another programming language such as C++ or Java. One example where I utilized the tapply function was in Question 2 of Project 11 where I was finding the sum of the grand total of orders for each vendor ID. 

```r
head(orders$vendor_id[year(orders$created_at)==2019])

head(orders$vendor_id[year(orders$created_at)==2020])

tapply(orders$vendor_id, year(orders$created_at), function(x){head(sort(table(x),decreasing=TRUE))})

df19<-subset(orders, year(created_at)==2019)

df20<-subset(orders, year(created_at)==2020)

tapply(df19$grand_total, df19$vendor_id,function(x){head(sort(sum(x),decreasing=TRUE))})

tapply(df20$grand_total, df20$vendor_id,function(x){head(sort(sum(x),decreasing=TRUE))})
```
