## Question 2

```r
first_digit <- function(number){
    if (number!=0){
        
        number <- abs(number)
    
        while(number >= 10){
            number <- number/10
        }
    
        while(number < 1){
            number <-number*10
        }
        number<-floor(number)
    }
    return (number)
}
```

```r
first_digit(0.0000546)
```

```r
first_digit(-10239485)
```

```r
first_digit(987654321)
```
