## Question 1

1.1) A container is an environment that allows code/software to be run in a common space without requiring a specific operating system to run the code. It enables users to store multiple apps in a location where the operating system is the same, so users do not have to worry about code not working on different operating systems. \
1.2) Docker images are essentially templates that are used to run a program in a container. You specify a type of code (Python, JavaScript, etc.) as a base image and then build your application around this base image. A registry is like a repository, as it is a centralized location where users can store Docker images by "uploading" (pushing) them to the registry for storage, and then "downloading" (pulling) them when they want to run their code.

```python
import requests

# Base URL for our API
base_url = "http://127.0.0.1:8000"
```

```python
#creating post requests to create items 
response1 = requests.post(f"{base_url}/items/1", params = {"value":"hello world"})
print("Created item 1: ", response1.json())

response2 = requests.post(f"{base_url}/items/2", params = {"value":"FastAPI Demo"})
print("Created item 2: ", response2.json())

response3 = requests.post(f"{base_url}/items/3", params = {"value":"Container API"})
print("Created item 3: ", response3.json())
```

```python
#creating get requests to read in data from db
for item_id in [1,2,3]:
    response = requests.get(f"{base_url}/items/{item_id}")
    print("Retrieved item: ", response.json())
```

```python
#creating an put request to update the data 
response = requests.put(f"{base_url}/items/2", params={"value":"Updated FastAPI Demo"})
print("Updated item 2: ", response.json())
```

```python
#creating a get request to ensure our update worked
response = requests.get(f"{base_url}/items/2")
print("Item 2 after update: ", response.json())
```

```python
#creating a delete request to delete an item from the db
response = requests.delete(f"{base_url}/items/3")
print("Deleted item 3: ", response.json())
```

```python
#attempting to read a deleted item to show error handling
response = requests.get(f"{base_url}/items/3")
print("Status Code: ", response.status_code)
print("Response: ", response.json())
```

```python
#using the post method to create my own item
response = requests.post(f"{base_url}/items/4", params={"value":"Testing POST Method"})
print("Added item: ", response.json())
```
