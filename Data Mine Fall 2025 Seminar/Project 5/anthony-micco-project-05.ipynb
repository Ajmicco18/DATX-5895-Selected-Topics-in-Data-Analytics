## Question 4

```python
# finding the average height of players with the last name 'Sanders'
```

```python
%%sql
SELECT *, AVG(height) AS avg_height FROM People WHERE nameLast = 'Sanders' GROUP BY nameLast;
```

```python
#finding the average weight of players grouped by their birth year and sorting the results by average weight 
```

```python
%%sql
SELECT *, AVG(weight) AS avg_weight FROM People GROUP BY birthYear ORDER BY avg_weight DESC LIMIT 10
```
