```python
%%bash
echo "apple banana apple orange banana apple" | awk '
{
    # Loop through each word in the input
    for (i = 1; i <= NF; i++) {
        # Increment the count for the word
        # $i is the current word
        # count[$i] is the count for the word
        # ++ is the increment operator
        count[$i]++
    }
}
END {
    print "Word counts:"
    for (word in count) {
        print word ": " count[word]
    }
}'
```

```python
# finding the average order amount by customer
```

```python
%%bash
cat /anvil/projects/tdm/data/restaurant/orders.csv | awk -F',' '
NR > 1 {  # Skip header row
    customer_id = $2
    order_amount = $4 + 0  # Convert string to number by adding 0
    if (customer_id != "" && order_amount > 0) { # ensure that the customer and order actually exist
        total_spent[customer_id] += order_amount
        order_count[customer_id]++
    }
}
END {
    print "Customer ID, Total Spent, Number of Orders, Average Order"
    for (customer in total_spent) {
        avg = total_spent[customer] / order_count[customer]
        printf "%s,%.2f,%d,%.2f\n", customer, total_spent[customer], order_count[customer], avg
    }
}' | sort -t, -k4,4nr | head -n 10
```

```python
%%bash
cat /anvil/projects/tdm/data/restaurant/orders.csv | awk -F',' '
NR > 1 {  # Skip header row
    vendor_id = $22
    discount_amt = $7+0
    if (vendor_id != "" && discount_amt > 0) {
        total_discount[vendor_id] += discount_amt
        order_count[vendor_id]++
    }
}
END {
    print "Vendor ID, Total Discount Amt, Order Count, Average Discount"
    for (vendor in order_count) {
        avg = total_discount[vendor] / order_count[vendor]
        printf "%s,%.2f,%d,%.2f\n", vendor, total_discount[vendor], order_count[vendor], avg
    }
}' | sort -t, -k4,4nr | head -n 10
```

```python
#finding the highest average vendor rating
```

```python
%%bash
cat /anvil/projects/tdm/data/restaurant/orders.csv | awk -F',' '
NR > 1 {  # Skip header row
    vendor_id = $22
    vendor_rating = $11+0
    if (vendor_id != "" && vendor_rating > 0) {
        total_rating[vendor_id] += vendor_rating
        order_count[vendor_id]++
    }
}
END {
    print "Vendor ID, Order Count, Average Rating"
    for (vendor in order_count) {
        avg = total_rating[vendor] / order_count[vendor]
        printf "%s,%d,%.2f\n", vendor,order_count[vendor], avg
    }
}' | sort -t, -k3,4nr | head -n 10
```
