```python
%%sql
CREATE TRIGGER IF NOT EXISTS update_yearID
AFTER INSERT ON Teams
FOR EACH ROW
BEGIN
    UPDATE Teams
    SET yearID = strftime('%Y', 'now')
    WHERE teamID = NEW.teamID;
END;
```

## Question 2

```python
# creating the new DeathRecords table
```

```python
%%sql
DROP TABLE IF EXISTS DeathRecords;

CREATE TABLE DeathRecords (
    playerID TEXT PRIMARY KEY,
    yearID INTEGER NOT NULL,
    monthID INTEGER NOT NULL,
    dayID INTEGER NOT NULL
)
```

```python
#create a trigger to update the People table when rows are inserted into DeathRecords table
```

```python
%%sql
CREATE TRIGGER IF NOT EXISTS update_People
AFTER INSERT ON DeathRecords
FOR EACH ROW
BEGIN
    UPDATE People
    SET deathYear = NEW.yearID,
    deathMonth = NEW.monthID,
    deathDay = NEW.dayID
    WHERE playerID = NEW.playerID;
END;
```

```python
#inserting a new player to see if the trigger works
```

```python
%%sql
INSERT INTO DeathRecords (playerID, yearID, monthID, dayID)
VALUES ('abreubo01', 2025, 1, 1);
```

```python
#checking to see if trigger worked
```

```python
%%sql
SELECT * FROM People WHERE playerID = 'abreubo01';
```

```python
#inserting multiple records into the table to see if the trigger functions for multiple values
```

```python
%%sql
INSERT INTO DeathRecords (playerID, yearID, monthID, dayID)
VALUES
('pintore01', 2025, 2, 1),
('hundlni01', 2025, 3, 1),
('detwiro01', 2025, 4, 1);
```

```python
#ensuring the trigger functioned properly
```

```python
%%sql
SELECT playerID, deathYear, deathMonth, deathDay FROM People
WHERE playerID IN ('pintore01', 'hundlni01', 'detwiro01');
```

```python
# creating the trigger to update the People table after a new player is added to the DeathRecords table
```

```python
%%sql
CREATE TRIGGER IF NOT EXISTS update_players_death
AFTER INSERT ON DeathRecords
FOR EACH ROW
BEGIN
    UPDATE People
    SET deathYear = NEW.yearID,
        deathMonth = NEW.monthID,
        deathDay = NEW.dayID
    WHERE playerID = NEW.playerID;
END;
```

```python
#creating the DeceasedPlayers view that returns the playerID, nameFirst, nameLast, deathYear, deathMonth, and deathDay for player with a nonNull deathYear
```

```python
%%sql
DROP VIEW IF EXISTS DeceasedPlayers;

CREATE VIEW IF NOT EXISTS DeceasedPlayers AS
SELECT playerID, nameFirst, nameLast, deathYear, deathMonth, deathDay
FROM People
WHERE deathYear IS NOT NULL;
```
