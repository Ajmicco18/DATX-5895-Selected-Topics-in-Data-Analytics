```python
#Joining the salaries and people tables as a temporary table
```

```python
%%sql
WITH joined_data AS (
    SELECT * FROM Salaries JOIN People ON People.playerID = Salaries.playerID
)

SELECT * FROM joined_data LIMIT 5;
```

```python
#Joining salaries and people tables, finding the average salary of peoples grouped by weight and sorted in descending order
```

```python
%%sql
WITH joined_data AS (
    SELECT * FROM Salaries JOIN People ON People.playerID = Salaries.playerID
)

SELECT *, AVG(salary) AS avg_salary FROM joined_data GROUP BY weight ORDER BY weight DESC LIMIT 10;
```

```python
#joining the batting and people tables and finding how many runs Babe Ruth scored during his career
```

```python
%%sql
WITH joined_data AS (
    SELECT * FROM People JOIN Batting ON People.playerID = Batting.playerID
)

SELECT SUM(R) FROM joined_data WHERE nameFirst = 'Babe' AND nameLast = 'Ruth'
```

```python
# writing a query to find the average number of runs scored by players in the 1980s
```

```python
%%sql
WITH joined_data AS (
    SELECT * FROM People JOIN Batting ON People.playerID = Batting.playerID
)

SELECT birthYear, AVG(R) AS avg_runs FROM joined_data WHERE nameLast NOT LIKE '%%' AND birthYear >= 1980 AND birthYear < 1990 GROUP BY birthYear ORDER BY birthYear DESC
```

## Question 4

```python
# query to find the players with the most runs scored in their careers
```

```python
%%sql

WITH joined_data AS (
    SELECT * FROM people JOIN batting ON people.playerID = batting.playerID
)

SELECT playerID,nameFirst,nameLast, SUM(R) AS totalRuns FROM joined_data WHERE nameFirst NOT LIKE '%%' AND nameLast NOT LIKE '%%' GROUP BY playerID ORDER BY totalRuns DESC LIMIT 10
```

```python
# query to find the players with the most hits in their careers
```

```python
%%sql

WITH joined_data AS (
    SELECT * FROM people JOIN batting ON people.playerID = batting.playerID
)

SELECT playerID,nameFirst,nameLast, SUM(H) AS totalHits FROM joined_data WHERE nameFirst NOT LIKE '%%' AND nameLast NOT LIKE '%%' GROUP BY playerID ORDER BY totalHits DESC LIMIT 10
```

```python
# reading in the CollegePlaying table 
```

```python
%%sql
SELECT * FROM CollegePlaying LIMIT 5;
```

```python
#joining the CollegePlaying and batting tables to find the most runs scored by Purdue University players 
```

```python
%%sql
WITH joined_data AS (
    SELECT CollegePlaying.playerID, Batting.R FROM CollegePlaying JOIN Batting ON CollegePlaying.playerID = Batting.playerID  WHERE CollegePlaying.schoolID = 'purdue'
)

SELECT *, SUM(R) AS careerRuns FROM joined_data GROUP BY playerID ORDER BY careerRuns DESC LIMIT 10
```

```python
#query to find the average number of runs scored by players from a university
```

```python
%%sql
WITH joined_data AS (
    SELECT * FROM CollegePlaying JOIN Batting ON CollegePlaying.playerID = Batting.playerID
)

SELECT schoolID, AVG(R) AS avgRuns FROM joined_data GROUP BY schoolID ORDER BY avgRuns DESC LIMIT 10
```
