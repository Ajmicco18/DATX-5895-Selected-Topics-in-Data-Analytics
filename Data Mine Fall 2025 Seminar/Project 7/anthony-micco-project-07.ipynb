2.3) The difference in the results between the two queries is that the inner join query returns only the rows that have matching values between their tables. In our case, the inner join query returned only the rows from the People table where the playerID column matched with the playerID column from the salaries table. Therefore, there are no null values in an inner join. In contrast, the left join returns ALL rows from the table you are joining on the left and then all the matching values of the table you are joining, therefore null values can occur where they are places where the two tables do not have rows containing the same values. 

## Question 3

```python
# using the example query to understand how the HAVING clause works
```

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary
FROM Salaries s
JOIN Teams t ON t.teamID = s.teamID AND t.yearID = s.yearID
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 5000000
ORDER BY avgSalary DESC
LIMIT 10;
```

```python
#writing a query to join the salaries, people and teams table and then group results by year and team and filter only the teams with avgSalary > 7.5 million
```

```python
%%sql
SELECT 
t.name AS teamName,
s.yearID,
AVG(s.salary) AS avgSalary
FROM Salaries s 
JOIN Teams t ON t.teamID = s.teamID AND t.yearID=s.yearID
GROUP BY t.teamID, s.yearID
HAVING avgSalary > 7500000
ORDER BY avgSalary 
```

## Question 4

```python
# using the multiple condition HAVING clause query example to examine functionality
```

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numPlayers
FROM Salaries s
JOIN Teams t ON t.teamID = s.teamID AND t.yearID = s.yearID
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 5000000 AND COUNT(s.playerID) > 20
ORDER BY avgSalary DESC
LIMIT 10;
```

```python
# query to find teams with average salary greater than 6.8, more than 20 players and the year being between 2010 and 2015 
```

```python
%%sql
SELECT 
t.name AS teamName,
s.yearID,
AVG(s.salary) AS avgSalary,
COUNT(s.playerID) AS totalPlayers
FROM Salaries s
JOIN Teams t ON t.teamID = s.TeamID AND t.yearID = s.yearID
WHERE s.yearID BETWEEN 2010 AND 2015
GROUP BY t.name, s.yearID
HAVING avgSalary > 6800000 AND totalPlayers > 20
ORDER BY avgSalary DESC
```

## Question 5

```python
# creating the payroll aggregation for each team in each year
```

```python
%%sql 
WITH payroll AS (
    SELECT
        teamID,
        yearID,
        SUM(salary) AS totalPayroll
    FROM Salaries
    GROUP BY teamID, yearID
)
SELECT * FROM payroll WHERE totalPayroll > 150000000 LIMIT 10
```

```python
#finding the max number of home runs for each player in each year 
```

```python
%%sql
WITH maxHR AS (
    SELECT 
        teamID,
        playerID,
        yearID,
        SUM(HR) AS totalHR
    FROM Batting
    GROUP BY playerID, teamID, yearID
)

SELECT teamID,playerID, yearID, MAX(totalHR) AS maxHR 
FROM maxHR 
GROUP BY teamID, yearID
HAVING maxHR >= 30
ORDER BY maxHR DESC LIMIT 20
```

```python
#joining the two aggregates and filtering to keep only qualifying team-seasons
```

```python
%%sql
WITH payroll AS (
    SELECT
        teamID,
        yearID,
        SUM(salary) AS totalPayroll
    FROM Salaries
    GROUP BY teamID, yearID
),
maxHR AS (
    SELECT 
        teamID,
        playerID,
        yearID,
        SUM(HR) AS totalHR
    FROM Batting
    GROUP BY playerID, teamID, yearID
)
SELECT 
    p.teamID,
    p.yearID,
    p.totalPayroll,
    m.playerID,
    m.totalHR
FROM payroll p
JOIN maxHR m ON m.teamID = p.teamID AND m.yearID = p.yearID
GROUP BY p.teamID, p.yearID
HAVING MAX(m.totalHR) >= 30 AND p.totalPayroll > 150000000
ORDER BY p.totalPayroll LIMIT 5
```
