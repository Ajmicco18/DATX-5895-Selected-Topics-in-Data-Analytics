```python
#finding the number of unique zip codes in each data set

#reading in map data from Tippecanoe County with Address Points
q1Data = gpd.read_file('/anvil/projects/tdm/data/tippecanoe/AddressPoints.shp')
pd.unique(q1Data["GEOZIP"])
```

```python
#reading in map data from Tippecanoe County with Parcel Data
q3Data = gpd.read_file('/anvil/projects/tdm/data/tippecanoe/Parcels.shp')
pd.unique(q3Data["PROP_ZIP"])
```

```python
#reading in the data from Tippecanoe County with Street Centerline data
q4Data = gpd.read_file('/anvil/projects/tdm/data/tippecanoe/StreetCenterlines.shp')
pd.unique(q4Data["ESRI_ZIP"])
```

I will use the data from question 4 to assign a different color to each zip code and visualize the street centerlines in each zip code

```python
#creating an array of unique zips to iterate through 
zips = pd.unique(q4Data["ESRI_ZIP"])
```

```python
#adding the mycolors column to the data frame
q4Data = q4Data.assign(mycolors = 'orange')
```

```python
#creating a function to generate a random color for each zip code

#importing random to randomly choose letters for hex color
import random

#defining function
def create_random_color():
    #creating a hex color using the characters 0-9 and a-f and joining them together to form a color using hex
    hex_color = '#' + ''.join([random.choice('0123456789abcdef') for j in range(6)])
    return hex_color
```

```python
#iterating through each zip code in the zips array and assigning a different color to each zip code
for zip in zips: 
    q4Data.loc[q4Data["ESRI_ZIP"] == zip, "mycolors"] = create_random_color()
```

```python
#ensuring the random colors generated properly
pd.unique(q4Data["mycolors"])
```

```python
q4Data.plot(color=q4Data["mycolors"])
```
